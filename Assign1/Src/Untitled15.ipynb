```python
import pandas as pd

# Loading the  raw data
raw_data = pd.read_csv('female_participants_data.csv')

raw_data.columns = ['Height', 'Weight', 'Age', 'Grip_strength', 'Frailty']

raw_data['Grip_strength'] *= 2.204

raw_data['Grip_strength'] = raw_data['Grip_strength'].round(2)

raw_data['Frailty'] = raw_data['Frailty'].str.strip().map({'Y': 1, 'N': 0})


print(raw_data.head())


raw_data.to_csv('cleaned_female_participants_data.csv', index=False, encoding='utf-8')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Loading the cleaned data
cleaned_data = pd.read_csv('cleaned_female_participants_data.csv')

# Visualization of the  height distribution
plt.hist(cleaned_data['Height'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Height (inches)')
plt.ylabel('Frequency')
plt.title('Distribution of Height')
plt.show()

#  relationship between the age and grip strength
plt.scatter(cleaned_data['Age'], cleaned_data['Grip_strength'], color='black')
plt.xlabel('Age (years)')
plt.ylabel('Grip Strength (pounds)')
plt.title('Relationship between Age and Grip Strength')
plt.show()

#  visualizing frailty distribution
frailty_counts = cleaned_data['Frailty'].value_counts()
plt.bar(frailty_counts.index, frailty_counts.values, color=['green', 'orange'])
plt.xlabel('Frailty')
plt.ylabel('Count')
plt.title('Distribution of Frailty')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```
